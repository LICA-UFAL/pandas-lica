# **.loc, um método muito importante**

- Pegar 1 linha
- Pegar linhas de acordo com alguma condição
- Pegar linhas e colunas específicas
- Pegar 1 valor específico

```python
# pegar uma linha
display(vendas_df.loc[1:5]) #da linha 1 ate a linha 5, caso queira so uma linha só por o número dela


# pegar Linhas que corresponde a uma condição
vendas_df['ID Loja'] == 'Iguatemi Esplanada'
vendas_ESPLANADA = vendas_df.loc[vendas_df['ID Loja'] == 'Iguatemi Esplanada' && vendas_df['Produto'] == 'Camiseta' ]

# pegar varias linhas e colunas usando .loc
#testing = vendas_df.loc[linhas,colunas]
vendas_ESPLANADA = vendas_df.loc[vendas_df['ID Loja'] == 'Iguatemi Esplanada', ['ID Loja','Produto', 'Quantidade']]
display(vendas_ESPLANADA)

#pegar 1 valor específico
print(vendas_df.loc[1, 'Produto'])
```

```python
vendas_ESPLANADA = vendas_df.loc[(vendas_df['ID Loja'] == 'Iguatemi Esplanada') & (vendas_df['Produto'] == 'Camiseta') ]
vendas_ESPLANADA
```

# **Adicionar 1 coluna**

```python
# a partir de uma coluna que existe
vendas_df['Comissão'] = vendas_df['Valor Final']* 0.05
display(vendas_df)

# criar uma coluna com valor padrão
vendas_df.loc[:, 'Imposto'] = 0
display(vendas_df)
```

# **Adicionar 1 linha**

# **Excluir linhas e colunas**

```python
vendas_df = vendas_df.drop("Imposto", axis=1) #AXIS=0 É O EIXO DA LINHA, AXIS=1 É O EIXO DA COLUNA, como "imposto" é coluna, coloco axis=1
display(vendas_df)
```

# **Valores vazios**

- Deletar linhas e colunas vazias
- Deletar linhas que possuem valores vazios
- Preencher valores vazios(média e ultimo valor)

```python
# deletando linhas e colunas completamente vazias
#vendas_df = vendas_df.dropna(how='all') # -> se rodar assim, exclui as linhas onde todos os valores são vazios
#vendas_df = vendas_df.dropna(how='all', axis=1) # -> se rodar com o parametro axis=1, exclui todas as colunas onde os valores são vazios

# deletando linhas que possuem pelo menos um valor vazio
#vendas_df = vendas_df.dropna() # -> se rodar assim, ele exclui as linhas que possuem pelo menos um valor vazio(forma padrão)

# preenchendo valores vazios
# preenchendo com a média da coluna
# vendas_df["Comissão"] = vendas_df["Comissão"].fillna(vendas_df["Comissão"].mean())
# display(vendas_df)

# preenchendo com o último valor
vendas_df = vendas_df.ffill()
display(vendas_df)
```

# **Calcular Indicadores**

- Groupby
- Value Counts


```python
transacoesporloja = vendas_df['ID Loja'].value_counts()
display(transacoesporloja)

faturamentoporloja = vendas_df[['Produto','Quantidade','Valor Final']].groupby('Produto').sum()
display(faturamentoporloja)
```

# **Mesclar 2 dataframes (procurando informaçoes de um dataframe em outro)**
